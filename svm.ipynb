# Image recognition with Support Vector Machines

## 目的

#### Scikit-learnのSVC関数を使用し顔画像を分類<br> 画素値を特徴量として使用し，識別データのクラスラベルを予測

## データセット

#### sklearn.datasetsよりfetch_olivetti_facesを使用<br> fetch_olivetti_facesには1人10枚×40人，計400枚の顔画像が含まれている．<br> 画像は異なる照明条件や表情で撮影した写真(笑顔/笑顔でない,目を開けている/目を閉じている,眼鏡がある/眼鏡がない)<br> 画像サイズ:64×64

```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
print faces.DESCR
```

#### facesには4つのデータが入っており，各次元の要素数は以下の数となる．<br> imagesには0~1の正規化された値が64×64入っており，それが画像の枚数である400個存在する．<br> dataはimagesの2，3次元目(64×64)を1次元に纏めたものになる．

```python
print faces.keys()
print faces.images.shape
print faces.data.shape
print faces.target.shape

print faces.target
```

#### SVMに適用させて良い結果を出すために，データを正規化する事が重要である．<br> ただし，今回使用するデータの値が以下より最大値1，最小値0な為正規化する必要がない．

```python
print np.max(faces.data)   
print np.min(faces.data)   
print np.mean(faces.data)
```

#### 学習を行う前に，使用するデータセットの画像を出力する関数を作成する<br> カラーマップは青成分の値が高められるグレースケールのカラーマップ(bone)を使用

```python
def print_faces(images,target,top_n):
    fig = plt.figure(figsize=(12,12))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1,hspace=0.05, wspace=0.05)
    for i in range(top_n):
        p=fig.add_subplot(20,20,i+1,xticks=[],yticks=[])
        p.imshow(images[i],cmap=plt.cm.bone)
        p.text(0,14,str(target[i]))
        p.text(0,60,str(i))
```

```python
print_faces(faces.images,faces.target,400)
```

## 学習

#### 関数はscikit-learnにあるSVC関数を使用する

```python
from sklearn.svm import SVC
```

#### SVCは分類のため使用され，重要なパラメータとして分類器で使用されるカーネル関数を選択するkernelパラメータがある．デフォルトでは，非線形問題をモデル化することが出来るRBFカーネルを使用している．今回は最も単純なカーネルである，線形カーネルを使用する．

```python
svc_1 = SVC(kernel='linear')
```

#### データセットを学習データと識別データの2つに分割する．この時，学習データは75%識別データは25%とする．

```python
from sklearn.cross_validation import train_test_split
X_train,X_test,y_train,y_test = train_test_split(faces.data,faces.target,test_size=0.25,random_state=0)
```

#### K分割交差検証を使用するための関数を定義

```python
from sklearn.cross_validation import cross_val_score,KFold
from scipy.stats import sem

def evaluate_cross_validation(clf,X,y,K):
    # create a k-fold croos validation iterator
    cv = KFold(len(y),K,shuffle=True,random_state=0)
    # by default the score used is the one returned by score
    scores=cross_val_score(clf,X,y,cv=cv)
    print scores
    print ("Mean score: {0:.3f}(+/-{1:.3f})").format(np.mean(scores),sem(scores))
```

#### 5分割する事で結果的に0.913という結果が得られた．<br> これによりデータが偏りなく分割できている事が確認できる．

```python
evaluate_cross_validation(svc_1,X_train,y_train,5)
```

#### 学習データで学習を行い，識別データでの性能を評価する関数を定義<br> この関数により，学習データと識別データに対する正解率<br> Classification Reportにより，適合率，再現率，F値，正解ラベルのデータ数<br> Confusion Matrixを出力する事ができる．

```python
from sklearn import metrics

def train_and_evaluate(clf,X_train,X_test,y_train,y_test):
    clf.fit(X_train,y_train)
    print "Accuracy on training set:"
    print clf.score(X_train,y_train)
    print "Accuracy on testing set:"
    print clf.score(X_test,y_test)
    
    y_pred = clf.predict(X_test)
    
    print y_test
    print y_pred
    print "Classsification Report:"
    print metrics.classification_report(y_test,y_pred)
    print "Confusion Matrix:"
    print metrics.confusion_matrix(y_test,y_pred)
```

```python
train_and_evaluate(svc_1,X_train,X_test,y_train,y_test)
```

#### 分類した際に，眼鏡のあるなしを考慮したい．<br> 眼鏡をかけている顔の番号を定義する．

```python
#the index ranges of images of people with glasses
glasses = [
      (10, 19), (30, 32), (37, 38), (50, 59), (63, 64),
      (69, 69), (120, 121), (124, 129), (130, 139), (160, 161),
      (164, 169), (180, 182), (185, 185), (189, 189), (190, 192),
      (194, 194), (196, 199), (260, 269), (270, 279), (300, 309),
      (330, 339), (358, 359), (360, 369)]
```

#### 400の顔画像に対して左上に1か0の数字を表示する．<br>眼鏡をかけている場合は1，眼鏡をかけていない場合は0を与える関数を定義

```python
def create_traget(segments):
    y=np.zeros(faces.target.shape[0])
    for (start,end) in segments:
        y[start:end + 1] =1
    return y
```

```python
target_glasses = create_traget(glasses)
```

#### 再び分割検証を用いて学習データと識別データに分ける

```python
X_train, X_test, y_train, y_test = train_test_split(faces.data, target_glasses, test_size=0.25, random_state=0)
```

#### 新しいSVC分類器を作成する

```python
svc_2 = SVC(kernel='linear')
```

#### 以下のコードにより，分割検証を行いパフォーマンスを調べる

```python
evaluate_cross_validation(svc_2,X_train,y_train,5)
```

#### 以下の結果より，分類器が眼鏡のあるなしを識別出来ている事が確認できる．

```python
train_and_evaluate(svc_2,X_train,X_test,y_train,y_test)
```

#### 最後に学習データを390，識別データを10として学習を行い，識別をする．<br> 識別データは眼鏡のあるなし両方が混ざったデータを使用する．

```python
X_test = faces.data[30:40]
y_test = target_glasses[30:40]
print y_test.shape[0]
select = np.ones(target_glasses.shape[0])
select[30:40]=0
X_train=faces.data[select==1]
y_train=target_glasses[select==1]
print y_train.shape[0]
svc_3 = SVC(kernel='linear')
train_and_evaluate(svc_3,X_train,X_test,y_train,y_test)
```

#### 結果は良いが1つ誤識別をおこしてしまった．<br> 間違って分類を行った画像を見てみる

```python
y_pred = svc_3.predict(X_test)
eval_faces=[np.reshape(a,(64,64))for a in X_test]
```

#### 下記の画像の8番目の画像を誤識別した．<br> 原因は眼鏡の境界線が分かりにくいのと目を閉じているのが原因

```python
print_faces(eval_faces, y_pred, 10)
```

#### 今回顔の分類に線形SVMを使用したが，たいていは最初の試行でこのような良い結果は得られない．<br> SVMは特定のケースでは異なるカーネルを利用できる．線形で結果が得られないならば，多項式やRBFカーネルを使える．<br> その他にはCやgammaパラメータを調整する事で，結果に影響を与える事ができる．
