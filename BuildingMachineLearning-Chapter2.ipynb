# 2章 実例を対象とした分類法入門

花の写真から品種を当てる，といった事をコンピュータにさせたい．機械学習の観点では，品種別に花の画像を用意してコンピュータに分類する方法を学習させる．品種を当てるためには学習したルールを使う事で分類を行う．これは，<b>クラス分類(classification)</b>または，<b>教師あり学習(supervised learning)</b>と呼ばれる．<br>
本章では，単純なアルゴリズムを用いて，サイズの小さいデータセットを対象にクラス分類を行う事で，基本的な原理を理解する事を目的とする．

## 2.1 アイリスデータセット

使用するデータセットの中身には，花に関する情報が入っている．アイリスにはいくつか品種があり，アイリスデータセットは3つの異なる品種のサンプルデータから構成されている．このデータセットは，サイズが小さいため(サンプル数:150，特徴量の数:4)データの可視化や分析が簡単に行える．<br>また，サンプルされたアイリスを対象に，下記の4つの要素がそれぞれ計測されている．<br>
<br>
<LI>がく片の長さ(Sepal length)</LI>
<LI>がく片の幅(Sepal width)</LI>
<LI>花弁の長さ(Petal length)</LI>
<LI>花弁の幅(Petal width)</LI>

計測されたデータは<b>特徴量(features)</b>と呼ばれる．今回の場合は次の問題を考える．<br>
外を歩いているとアイリスという花を見つけた．花の品種を知るためにサイズを計測したが，アイリスの品種を出来るだけ正確に予測するにはどうすればよいか？<br>
<br>
これは，教師あり学習またはクラス分類に関する問題となる．与えられたサンプルデータにはラベル付けがされており，そのデータから分類を行うための方法を設計する事ができる．

### 2.1.1 データの可視化

今回扱うデータは非常に小さいので，全ての点を描画する事が出来る．また，2次元からなるグラフを描画する場合，軸として採用する特徴量の組み合わせについても，全ての組み合わせで描画できる．表示したグラフを見ると，花の種類が3つある事が確認でき，それぞれ以下のように分類される．<br>
<br>
<LI>三角形:Setosa</LI>
<LI>丸印:Versicolor</LI>
<LI>バツ印:Virginica</LI>

```python
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

#sklearnからload_iris関数を用いて，データをロードする
data = load_iris()

features = data['data']
features_names = data['feature_names']
target = data['target']
labels = data['target_names']

pairs = [(0,1),(0,2),(0,3),(1,2),(1,3),(2,3)]
fig,axes = plt.subplots(2,3,figsize=(12,8))

print fig
for i, (p0,p1) in enumerate(pairs):
    ax = axes.flat[i]
    for t, marker, c in zip(range(3), ">ox", "rgb"):
        #クラスごとに色の異なるマーカでプロットする
        ax.scatter(features[target == t,p0],features[target == t,p1],marker=marker,c=c,s=40)
    ax.set_xlabel(features_names[p0])
    ax.set_ylabel(features_names[p1])
    #ax.set_xticks([])
    ax.set_yticks([])
```

### 2.1.2 分類モデルの作成

今回の目的が花を3つのグループに分類することであれば，いくつか方法を思いつくだろう．<br>
たとえば，"花弁の長さ(Petal length)"を用いる事で，Setosaという品種のアイリスと他の品種を見分ける事ができそうである．<br>
以下のコードで，その境界線を見つける事ができる．

```python
target_names = data['target_names']
labels = target_names[target]
plength = features[:, 2]

is_setosa = (labels == 'setosa')
#setosaの花弁の長さの最大値をmax_setosaに格納
max_setosa = plength[is_setosa].max()
#他2種の花弁の長さの最小値をmin_non_setosaに格納
min_non_setosa = plength[~is_setosa].min()

print('Maximum of setosa: {0}.'.format(max_setosa))
print('Minimum of others: {0}.'.format(min_non_setosa))
```

<br>
上記のコードを実行すると，setosaの花弁の長さの最大値は1.9であり，他の2品種の花弁の長さの最小値は3.0であるという事がわかる．<br>
そのため，次のような単純なモデルを作成する事ができる．<br>
それは「もし花弁の長さが2より小さければ，それはSetosaという品種のアイリスであり，それ以外であれば，それはVirginicaかVersicolorという品種のアイリスである」というモデルである．<br>
以下にこのモデルを関数として定義する．

```python
def apply_model(example):
    if example[2] < 2 : print 'Iris Setosa'
    else: print 'Iris Virginica or Iris Versicolor'
```


これは本章で最初に設計したモデルとなる．単純だが，Setosaとその他のアイリスを誤り無く分類する事ができる．<br>
ここで行った事は1つの次元(特徴量)において閾値を定めただけだ．そのために，グラフを見ながら視覚的に分類に適している特徴量を見つけ，簡単な計算を行った．機械学習は以上のことを代わって行ってくれる．<br>
&nbsp;&nbsp;&nbsp;&nbsp;Setosaとその他のアイリスを分類することは簡単であった．しかし，VirginicaとVersicolorについては，分類のための最適な閾値は簡単に見つける事はできない事が見て分かる．しかし，ここではできるだけ最前の方法で分類できないか考える．そこで，次のような簡単な計算から始める事にする．


```python
#まず，Setosa以外の特徴量とラベルを選択
setosa = (labels == 'setosa')
features = features[~is_setosa]
labels = labels[~is_setosa]
virginica = (labels == 'virginica')
```

上記のコードでは，Numpyに関連する演算を多く使用している．is_setosaはブーリアン型(真理値の真か偽の2値をとる)の配列であり，これを用いてSetosa以外の2品種のアイリスだけからなる配列(featuresとlabels)を生成できる．最後に，ラベルについての比較を行い，virginicaというブーリアン型配列を新たに作成する．<br>
&nbsp;&nbsp;&nbsp;&nbsp;次に，全ての特徴量とその閾値の組み合わせについての正解率(accuracy)を計算し，正解率の最も高い組み合わせを見つける．ここでの正解率とは，モデルが正しく分類できた割合を示す．また，閾値の候補は訓練データの各特徴量の値を用いることにする．

```python
best_acc = -1.0
for fi in range(features.shape[1]):
    #各特徴量ごとに閾値の候補を生成する
    thresh = features[:, fi].copy()
    thresh.sort()
    #全ての閾値でテストする
    for t in thresh:
        pred = (features[:, fi] > t)
        acc = (pred == virginica).mean()
        if acc > best_acc:
            best_acc = acc
            best_fi = fi
            best_t = t
print('Best cut is {0} on feature {1}, which achieves accuracy of {2:.1%}.'.format(
    best_t, best_fi, best_acc))
```

<br>
上記のコード中のpredという変数は，各サンプルデータがvirginicaであるかどうかを示す予測結果であり，ブーリアン型の配列となる．そして，このpred配列のTrueである要素のラベル名を正解のラベル名であるvirginicaと比較を行う．<br>
&nbsp;&nbsp;&nbsp;&nbsp;また，ここではブーリアン型配列の平均(mean)をとることで，ブーリアン型配列のTrueの割合を求める事ができる．このTrueの割合が，正解率の値でもあるため簡単な記述で正解率の計算が行える．<br>
&nbsp;&nbsp;&nbsp;&nbsp;forループの最後では，全ての閾値と特徴量の組み合わせで評価が行われ，より適した値がbest_fiとbest_tに格納される．以下にこのモデルを使用するための関数を定義する．

```python
def apply_model(example):
    if example[best_fi] > best_t: print 'virginica'
    else:print 'versicolor'
```

全てのデータを対象に上記のコードを実行すると，最適なモデルは「花弁の幅」について分類するモデルであることがわかる．<br>
<br>
下記のコードによって表示される図を見ると，2つの領域に分けられる．一つは白であり，もう一つはグレイの領域となる．<br>白の領域にあるものはすべてVirginicaであり，それ以外はVersicolorの品種に分類されるということを表している．<br>
&nbsp;&nbsp;&nbsp;&nbsp;閾値を用いるモデルでは，決定境界(decision boundray)は常に一つの軸に対して平行な直線になる．今回の図では決定境界が示されており，各点は白かグレイ領域のどちらかに属する．また，同じ正解率で分類できる閾値はもう一つ存在し，そちらはより間隔の短い点線で示している．どちらの閾値を用いるかは任意に選択できる．

```python
COLOUR_FIGURE = False

from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
data = load_iris()
features = data['data']
feature_names = data['feature_names']
species = data['target_names'][data['target']]

setosa = (species == 'setosa')
features = features[~setosa]
species = species[~setosa]
virginica = species == 'virginica'

#閾値の値をtに代入する
t = 1.75
#使用する特徴量の次元数を指定
p0, p1 = 3, 2

if COLOUR_FIGURE:
    area1c = (1., .8, .8)
    area2c = (.8, .8, 1.)
else:
    area1c = (1., 1, 1)
    area2c = (.7, .7, .7)

x0, x1 = [features[:, p0].min() * .9, features[:, p0].max() * 1.1]
y0, y1 = [features[:, p1].min() * .9, features[:, p1].max() * 1.1]

plt.figure(figsize=(8,6))
plt.fill_between([t, x1], [y0, y0], [y1, y1], color=area2c)
plt.fill_between([x0, t], [y0, y0], [y1, y1], color=area1c)
plt.plot([t, t], [y0, y1], 'k--', lw=2)
plt.plot([t - .1, t - .1], [y0, y1], 'k:', lw=2)
plt.scatter(features[virginica, p0],
            features[virginica, p1], c='b', marker='o')
plt.scatter(features[~virginica, p0],
            features[~virginica, p1], c='r', marker='x')
plt.ylim(y0, y1)
plt.xlim(x0, x1)
plt.xlabel(feature_names[p0])
plt.ylabel(feature_names[p1])
```

### 交差検定で評価を行う

前述で定義したモデルを使用すると訓練データの94%を正しく分類する事ができる．<br>
しかし，評価を行う為に用いたデータは，閾値を決定するために用いた訓練データと同じである．学習モデルは，学習に使用していないデータを用いて評価を行う場合よりも，学習に使用したデータを用いて評価を行う場合のほうが良い結果を示す．<br>
&nbsp;&nbsp;&nbsp;&nbsp;今回は，「新しいデータに対しての汎用的な能力」について，正しい評価を行う事を目的とする．ここで，新しい未知のデータに対して正しく分類できる能力を<b>汎化性能(generalization ability)</b>と言う．<br>
&nbsp;&nbsp;&nbsp;&nbsp;この汎化性能について評価を行うためには，訓練に使用しなかったデータを用いらなければならない．そのため，ホールドアウトデータ(訓練データから一部を除外するデータ)を用いて，より正当な評価を行うことにする．<br>
<br>
まず，先ほど正解率を計算するためのプログラムを関数化する．

```python
def learn_model(features, labels):
    best_acc = -1.0
    for fi in range(features.shape[1]):
        thresh = features[:, fi].copy()
        thresh.sort()
        for t in thresh:
            pred = (features[:, fi] > t)
            acc = (pred == labels).mean()
            if acc > best_acc:
                best_acc = acc
                best_fi = fi
                best_t = t
    return best_t, best_fi


def apply_model(features, model):
    t, fi = model
    return features[:, fi] > t


def accuracy(features, labels, model):
    preds = apply_model(features, model)
    return np.mean(preds == labels)
```

次にデータを分割し，評価を行う．使用するのはsetosaを抜いた2品種(versicolor,virginica)のデータであり，学習データ:50，評価データ:50に分割する．

```python
data = load_iris()
features = data['data']
labels = data['target_names'][data['target']]

setosa = (labels == 'setosa')
features = features[~setosa]
labels = labels[~setosa]
virginica = (labels == 'virginica')

testing = np.tile([True, False], 50)
training = ~testing

model = learn_model(features[training], virginica[training])
train_error = accuracy(features[training], virginica[training], model)
test_error = accuracy(features[testing], virginica[testing], model)
#Nは評価データの数を表す
print('''\
Training error was {0:.1%}.
Testing error was {1:.1%} (N = {2}).
'''.format(train_error, test_error, testing.sum()))
```

<br>
プログラムを実行した結果を見ると，評価データを用いて評価を行った場合の方が，学習データを用いた場合よりも正解率が低くなった．<br>
上記のコードでは，学習データの中から一部データを取り除き，取り除いたデータを評価データとして用いた．ただし，この手法の問題点は，学習用のデータが少なくなる事である．その一方で評価データを減らす事も出来るが，そうする場合評価するためのデータ数が少なくなる事を意味する．理想的には，全てのデータを学習に使い，全てのデータを評価にも使いたい．<br>
全てのデータを学習に使い，全てのデータを評価にも使う事は出来ないが，<b>交差検定(cross-validation)</b>という手法を用いる事で，似たような事ができる．極端な例として，データを一つ省いて行う交差検定がある．これは<b>leave-one-out法</b>とも呼ばれる．このleave-one-out法は，学習データから一つデータを選び，それを取り除いたデータ集合でモデルを学習させる．そして，訓練に用いなかったデータ(この場合は一つのデータ)をそのモデルが正しく分類できるか評価する．

```python
def load_dataset(dataset_name):
    '''
    data,labels = load_dataset(dataset_name)

    Load a given dataset

    Returns
    -------
    data : numpy ndarray
    labels : list of str
    '''
    data = []
    labels = []
    with open('./data/{0}.tsv'.format(dataset_name)) as ifile:
        for line in ifile:
            tokens = line.strip().split('\t')
            data.append([float(tk) for tk in tokens[:-1]])
            labels.append(tokens[-1])
    data = np.array(data)
    labels = np.array(labels)
    return data, labels
```

下記のコードはleave-one-out法を行うプログラムであり，使用するデータは上記のコードで"seeds.csv"というデータを読み込む．今回は10回行い，各評価値の合計を10で割る事により評価を行う．

```python
features, labels = load_dataset('seeds')
labels = (labels == 'Canadian')
error = 0.0
for fold in range(10):
    training = np.ones(len(features), bool)
    training[fold::10] = 0
    testing = ~training
    model = learn_model(features[training], labels[training])
    test_error = accuracy(features[testing], labels[testing], model)
    error += test_error

error /= 10.0

print('Ten fold cross-validated error was {0:.1%}.'.format(error))
```

<br>
&nbsp;&nbsp;leave-one-out法の問題は，今回の場合，本来の100倍以上の計算をしなければならない事である．実際，全てのデータを対象としているため，データが増えるにしたがって，処理時間も増加する．<br>
&nbsp;&nbsp;&nbsp;&nbsp;別の方法で，<b>K-分割交差検定(K-fold Cross Validation)</b>という手法を用いることにより，leave-one-out法の処理コストを数分の1に減らすことができる．K-分割交差検定のKの値はたとえば5など，小さい数字が一般的である．もし5分割交差検定を行うとすると，データセットを5つのブロックに分割する．そして，1つのブロックを除いたデータ集合を訓練データとして用い，除いたブロックをテスト用のデータとして用いる．同じことを各ブロック対象に，全部で5回行う．<br>
&nbsp;&nbsp;&nbsp;&nbsp;以上の事を行うためのコードは，上記のコードと殆ど同じになる．ただし，上記のコードではデータを1つだけ除いたのに対し，今回は全体の20%のデータを除くこととなる．学習モデルは除外したデータをテストデータとして用いて評価を行う．そして，各回で算出した誤差を平均して最終結果を出力する．

```python
from IPython.display import Image
i = Image(filename='./data/K-fold.jpg')
display(i)
```

<br>
&nbsp;&nbsp;&nbsp;&nbsp;上の図は，5分割交差検定のプロセスを表している．データセットは5つのグループに分割されている．それぞれの分割したブロック(foldともよばれる)で，1つのブロックをテスト用データに，残りの4つを訓練用データに用いる．分割する数は自分で決めることができる．一般的には5から10ぐらいを設定すると良い(それぞれ全体の80%または90%を訓練データに使用することになる)．<br>
極端な例としては，データ数と同じ数だけ分割すこともできる．その場合はleave-one-out法を行っていることとなる．<br>

データを分割するときには，データに偏りが生じないように注意しなければならない．例えば，分割したデータの中には同じクラスしか存在しない，というようなことが起こる可能性がある．それでは，正しい評価を行う事ができない．ここでは，機械学習ライブラリが代わりにやってくれる為，データを分割する方法の詳細には触れない．<br>
交差検定を行うことで，モデルを一つだけではなく複数作ったことになる．ただし，今後新しいデータに対して用いるモデルとして，どのモデルを選択すべきか分からない．<br>
最も単純な方法は，交差検定で作ったモデルは使わずに，全ての訓練データを用いて学習したモデルを用いることである．交差検定は，最終的なモデルを作成するためではなく，対象のモデルがどれだけ汎用的な能力を持つか，ということを判断するために用いる．<br>

## 2.2 さらに複雑なクラス分類器の作成

&nbsp;&nbsp;&nbsp;&nbsp;前節では単純なモデルを用いた．そのモデルとは，一つの次元(特徴量)にたいして閾値を用いるものであった．この他にも様々なタイプのモデルを見ていく．ただし，各モデルの詳細について万遍なく説明はしない．まず，クラス分類モデルを構成している要素として，次の3つの分ける事ができる．<br
>
<LI>モデル構造</LI>
&nbsp;&nbsp;&nbsp;&nbsp;ここでは，ある一つの特徴量に対する閾値を用いた．<br>
<LI>探索アルゴリズム</LI>
&nbsp;&nbsp;&nbsp;&nbsp;ここでは，特徴量と閾値の全ての組み合わせを試行し，最適な組み合わせを選んだ．<br>
<LI>損失関数</LI>
&nbsp;&nbsp;&nbsp;&nbsp;損失関数を用いることで，より悪くないモデルを選択できる．訓練誤差を用い，これをひとつの到達点とすることで，最適な正解率に到達できたか，判断の一つとして考えることができる．一般的に，損失関数を最小にするような手法について考えることとなる．<br>

&nbsp;&nbsp;&nbsp;&nbsp;上の3つの要素を変更することで，異なる結果が得られる．たとえば，訓練誤差を最小にする閾値を求めるような場合，各特徴量について次の3つの値だけテストすることで，閾値を求めることができる．3つの値とは，「特徴量の平均値・平均値に標準偏差を1つ足した値・平均値に標準偏差を1つ引いた値」となる．これは処理が重い場合に効果的である(または，データ数が億を超える場合など)．<br>
&nbsp;&nbsp;&nbsp;&nbsp;先ほど用いた，全ての特徴量と閾値の組み合わせをチェックする「力まかせ探索」は現実的ではない場合がある．そのときは，今述べたような近似的な手法を用いなければならない．<br>
また，別の損失関数を用いることができる．損失関数が変われば，あるタイプの誤差が大きくなる，ということもあり得る．<br>
閾値による単純な分類手法は，とても限定されたものである．単純なモデルでも上手くいく場合は，今回のアイリスデータセットのような単純なケースに限られるだろう．<br>

## 2.3 より複雑なデータセットとクラス分類

&nbsp;&nbsp;&nbsp;&nbsp;ここでは，先ほどの例よりも少しだけ複雑なデータセットを扱う．その複雑さに対応するため，新しい分類アルゴリズムと参考になるアイデアをいくつか紹介する．

### 2.3.1 種(たね)データセットを学習する

&nbsp;&nbsp;&nbsp;&nbsp;ここでは，農業に関するデータセットを見ていく．データ数は少ないが，アイリスデータセットで行ったような全ての特徴量の組み合わせをプロットするには，特徴量の数が多すぎる．ここで扱うデータは，小麦の種に関する計測データであり，次に示す7つの特徴量が計測されている．<br>
<br>
<LI>面積(A)</LI>
<LI>周囲の長さ(P)</LI>
<LI>密集(C=4π A/P<sup>2</sup>)</LI>
<LI>長さ</LI>
<LI>幅</LI>
<LI>非対象係数</LI>
<LI>殻溝の長さ</LI>

データセットには3つの品種の小麦が含まれており，カナダ産，コマ産，ロサ産である．これまで通り，形状に関する計測データから，小麦の品種を分類することが目的となる．<br>

### 2.3.2 特徴量と特徴エンジニアリング

小麦データセットの特徴量である「密集度」は，新たに計測されたデータではなく，他の特徴量である「面積」と「周囲長さ」から計算されたものとなる．<br>
これは一般的に，<b>特徴エンジニアリング(feature engineering)</b>と呼ばれるテーマで行われる．機械学習において，特徴エンジニアリングよりアルゴリズムのほうが注目されがちだが，システム性能に大きな影響を及ぼすのは特徴エンジニアリングのほうである．<br>
&nbsp;&nbsp;&nbsp;&nbsp;ここでは，小麦のデータを集めた研究者が「密集度」という小麦の形状に関する特徴量を計算した．この特徴量は，同じ形状の小麦で大きさが2倍違っても同じ値になる．しかし，細長い小麦の場合は密集度は0に近づく一方で，丸みのある小麦の場合は1に近づく．<br>
<br>

学習において，良い特徴量を選択する事は大切である．ここでいう良い特徴量とは，次の2つの条件を同時に満たす特徴量であると言える．
<br>
<LI>重要なことには敏感に反応すること</LI>
<LI>重要ではないことには反応を示さないこと</LI>

たとえば，「密集度」については，小麦全体の大きさが変わっても密集度は変化しないが，形状が変わると密集度は変化する．実際のところ，今述べた2つの条件を完全に満たす特徴量を求めることは難しい．しかし，その理想に近づく特徴量を探索することは重要である．<br>
&nbsp;&nbsp;&nbsp;&nbsp;良い特徴量があるとすると，最適な特徴量を自動で選び出すことができるか，という問題が生じる．この問題は特徴選択(feature selection)として知られており，多くの方法が提案されている．しかし，実際の問題では，単純な方法が一番上手くいく．ここで扱うような小さい問題については，特徴選択を行ってもあまり意味がない．特徴選択を用いるのは，何千もの特徴量がある場合であり，そのような場合は有用でない特徴量を用いないことで，それ以降の処理を速くすることができる．

### 2.3.3 最近傍法

この小麦データセットを対象に，先ほどのアイリスデータセットで使用した手法を用いて，2つのクラスに分類できるか試してみることは可能だが，上手く分けることは出来ないだろう．そこで，ここでは新しい分類手法である<b>最近傍法(Nearest neighbor classification)</b>を用いる．<br>
&nbsp;&nbsp;&nbsp;&nbsp;はじめに，各データが，その特徴ベクトルによって表現されていると考える(数学的にはN次元空間の1点にデータが存在している考える)．そうすると，データ間の距離を計算することができる．距離を計算する方法(ユークリッド距離，マンハッタン距離など)にはいくつかあるが，その中から好きなものを選ぶことができる．たとえば，以下のように計算ができる．

```python
def distance(p0,p1):
    u'ユークリッド距離を計算する'
    return np.sum((p0-p1)**2)
```

ここでは，次の単純なルールを用いて分類を行うことにする．そのルールとは，新しいデータが与えられた場合，そのデータに最も近い点(最近傍点)をデータセットから探索し，その最近傍点のラベルを結果として返す関数になる．

```python
def nn_classify(training_set,training_labels,new_example):
    dists = np.array([distance(t,new_example)
        for t in training_set])
    nearest = dists.argmin()
    return training_labels[nearest]
```

この場合，ここで使用するモデルは学習データとそのラベルを全て保持する．そして，新しいデータが入力されたとき，つまり，分類時に全ての計算を行う．より良い実装方法は，最近傍の探索を高速に行う為に，学習時に特徴量をインデックス化することになる．しかし，そのためには複雑なアルゴリズムを用いらなければならない．<br>
&nbsp;&nbsp;&nbsp;&nbsp;このモデルは，学習データに対して誤り無く完璧に分類することができる．なぜなら，各データに一番近いデータはそれ自身であるため，そのラベルは必ず一致する(ただし，同じ値の特徴量を持つデータが複数存在し，そのラベルが事ある場合は，分類を誤る事がある)そのため，交差検定を用いて評価を行う事は必要不可欠となる．<br>
<br>
&nbsp;&nbsp;&nbsp;&nbsp;このモデルに対して，10分割の交差検定を用いて評価を行うと，88%の正解率となる．交差検定を用いて算出した正解率は，訓練データだけで評価したときの正解率より低くなる．しかし，交差検定の正解率のほうがモデルの性能についてより信頼ができる．<br>
次に決定境界について調査を行う．プロットするために，2次元でデータを表現するように単純化しなければならない．ここでは，特徴量の中で，面積と密集度の２つを軸にとってグラフ化をする．

```python
def fit_model(k, features, labels):
    '''Learn a k-nn model'''
    # There is no model in k-nn, just a copy of the inputs
    return k, features.copy(), labels.copy()


def plurality(xs):
    '''Find the most common element in a collection'''
    from collections import defaultdict
    counts = defaultdict(int)
    for x in xs:
        counts[x] += 1
    maxv = max(counts.values())
    for k, v in counts.items():
        if v == maxv:
            return k

# This function was called ``apply_model`` in the first edition
def predict(features, model):
    '''Apply k-nn model'''
    k, train_feats, labels = model
    results = []
    for f in features:
        label_dist = []
        # Compute all distances:
        for t, ell in zip(train_feats, labels):
            label_dist.append((np.linalg.norm(f - t), ell))
        label_dist.sort(key=lambda d_ell: d_ell[0])
        label_dist = label_dist[:k]
        results.append(plurality([ell for _, ell in label_dist]))
    return np.array(results)


def accuracy(features, labels, model):
    preds = predict(features, model)
    return np.mean(preds == labels)
```

```python
COLOUR_FIGURE = False

from matplotlib.colors import ListedColormap

feature_names = [
    'area',
    'perimeter',
    'compactness',
    'length of kernel',
    'width of kernel',
    'asymmetry coefficien',
    'length of kernel groove',
]


def plot_decision(features, labels):
    '''Plots decision boundary for KNN

    Parameters
    ----------
    features : ndarray
    labels : sequence

    Returns
    -------
    fig : Matplotlib Figure
    ax  : Matplotlib Axes
    '''
    y0, y1 = features[:, 2].min() * .9, features[:, 2].max() * 1.1
    x0, x1 = features[:, 0].min() * .9, features[:, 0].max() * 1.1
    X = np.linspace(x0, x1, 100)
    Y = np.linspace(y0, y1, 100)
    X, Y = np.meshgrid(X, Y)
    model = fit_model(1, features[:, (0, 2)], np.array(labels))
    C = predict(
        np.vstack([X.ravel(), Y.ravel()]).T, model).reshape(X.shape)
    if COLOUR_FIGURE:
        cmap = ListedColormap([(1., .6, .6), (.6, 1., .6), (.6, .6, 1.)])
    else:
        cmap = ListedColormap([(1., 1., 1.), (.2, .2, .2), (.6, .6, .6)])
    fig,ax = plt.subplots(figsize=(8,6))
    ax.set_xlim(x0, x1)
    ax.set_ylim(y0, y1)
    ax.set_xlabel(feature_names[0])
    ax.set_ylabel(feature_names[2])
    ax.pcolormesh(X, Y, C, cmap=cmap)
    if COLOUR_FIGURE:
        cmap = ListedColormap([(1., .0, .0), (.0, 1., .0), (.0, .0, 1.)])
        ax.scatter(features[:, 0], features[:, 2], c=labels, cmap=cmap)
    else:
        for lab, ma in zip(range(3), "Do^"):
            ax.plot(features[labels == lab, 0], features[
                     labels == lab, 2], ma, c=(1., 1., 1.))
    return fig,ax
features, labels = load_dataset('seeds')
names = sorted(set(labels))
labels = np.array([names.index(ell) for ell in labels])
fig,ax = plot_decision(features, labels)
```

上記のコードで実行された結果では，特徴量の形や領域の色は以下のように対応している．図では，それぞれの領域が水平方向に広がっているが，これはx軸(面積)が10から22までの範囲であるのに対して，y軸(密集度)が0.75から1.0までの範囲であるからである．そのため，x軸での変化はy軸での変化よりかなり大きくなっている．つまり，先ほどの距離を求める関数に従って計算する場合，ほとんどの場面で，x軸だけを考慮していることとなる．<br>
<br>
<LI>カナダ産(小麦):ひし形，領域:白</LI>
<LI>コマ産(小麦):円形，領域:黒</LI>
<LI>ロサ産(小麦):三角形，領域:グレイ</LI>



ここでは，それぞれの軸を共通のスケールにするため，正規化(normalize)する必要がある．そのための手法は数多くあるが単純な方法は，Zスコアを用いて正規化することである．Zスコアとは，平均からどれだけ離れているかということについて，標準偏差を基準として，求められる値になる．Zスコアが0のときは，そのデータが平均であり，正の値のときは平均より上であり，負の値のときは平均より下であることを意味する．<br>
今，全ての特徴量について正規化を行ったとして，そのデータで最近傍法を用いると，94%の正解率を得ることができる．もう一度，二次元空間での決定領域について見る為にグラフ化をする．

```python
features -= features.mean(0)
features /= features.std(0)
fig,ax = plot_decision(features, labels)
```

今度は，境界部分がより複雑になっているのが分かる．また，どちらの軸でも値が変われば，領域が変動するのがわかる．特徴量を全て用いると，以上のことを7次元空間で行うことになる．7次元空間を可視化することは難しいが，同じ原理が適用されることには変わりない．つまり，正規化を行うことで，それぞれの特徴量の重要性を圴一にできるということになる．正規化を行わなければ，特定の特徴量だけに結果を左右されることになる．<br>
最近傍法は単純な手法だが，それで十分な場合もある．また，最近傍法により一般化したk近傍法(k-nearest neighbor classifier)という手法もある．k近傍法は，最近傍点だけでなく，k個の近傍点を対象とする．k個の近傍点のラベルをそれぞれ投票して，一番多くのラベルを結果とする．一般的に，kの値は5など，小さな値を用いる．しかし，データセットが大きい場合などは，kを大きな値に設定することもできる．

## 2.4 二値分類と多項分類

本章で初めに紹介した分類器は，閾値によって分類を行う単純な分類器であった．これは，データの値が閾値より大きいかどうかによって，あるクラスかそれ以外かに分類した．そのため，これは<b>二項分類(binary classification)</b>と呼ばれる．また，二つ目の分類器は最近傍法という手法を用いた．この手法は複数あるクラスの中からどれか一つを結果として出力する．そのため，これは<b>多項分類(multiclass classification)</b>と呼ばれる．<br>
一般的に，二項分類を行う手法を設計するほうが，多項分類よりも単純である．しかし，多項分類を行う場合でも，その問題を二項分類からなる問題に分解することができる．前節ではアイリスデータセットで分類を行うときに，この考え方を採用した．
