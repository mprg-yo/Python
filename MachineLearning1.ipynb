# 1章 実例を対象とした分類法(p27~32)

花の写真から品種を当てる，といった事をコンピュータにさせたい．機械学習の観点では，品種別に花の画像を用意してコンピュータに分類する方法を学習させる．品種を当てるためには学習したルールを使う事で分類を行う．これは，<b>クラス分類(classification)</b>または，<b>教師あり学習(supervised learning)</b>と呼ばれる．<br>
この章では，単純なアルゴリズムを用いて，サイズの小さいデータセットを対象にクラス分類を行う事で，基本的な原理を理解する事を目的とする．

## 1.1 アイリスデータセット

使用するデータセットの中身には，アイリスという花に関する情報が入っている．アイリスにはいくつか品種があり，アイリスデータセットは3品種のサンプルデータから構成されている．このデータセットは，サイズが小さいため(サンプル数:150，特徴量の数:4)データの可視化や分析を簡単に行う事が可能である．<br>また，サンプルされたアイリスを対象に，下記の4つの要素がそれぞれ計測されている．<br>
<br>
<LI>がく片の長さ(Sepal length)</LI>
<LI>がく片の幅(Sepal width)</LI>
<LI>花弁の長さ(Petal length)</LI>
<LI>花弁の幅(Petal width)</LI>

計測されたデータは<b>特徴量(features)</b>と呼ばれる．今回の場合は次の問題を考える．<br>
「外を歩いているとアイリスという花を見つけた．花の品種を知るためにサイズを計測したが，アイリスの品種を出来るだけ正確に予測するにはどうすればよいか？」<br>
<br>
これは，教師あり学習またはクラス分類に関する問題となる．与えられたサンプルデータにはラベル付けがされており，そのデータから分類を行うための方法を設計する事ができる．

### 1.1.1 データの可視化

今回扱うデータは非常に小さいので，全ての点を描画する事が出来る．また，2次元からなるグラフを描画する場合，軸として採用する特徴量の組み合わせについても，全ての組み合わせで描画できる．表示したグラフを見ると，花の種類が3つある事が確認でき，それぞれ以下のように分類される．<br>
<br>
<LI>三角形(赤):Setosa</LI>
<LI>丸印(緑):Versicolor</LI>
<LI>バツ印(青):Virginica</LI>

```python
#使用するライブラリを定義
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

#sklearnからload_iris関数を用いて，データをロードする
data = load_iris()

#dataの中身を特徴量ごとに変数に格納
features = data['data']
features_names = data['feature_names']
target = data['target']
labels = data['target_names']

#生成するウィンドウの条件を定義
pairs = [(0,1),(0,2),(0,3),(1,2),(1,3),(2,3)]
fig,axes = plt.subplots(2,3,figsize=(12,8))
print fig

#featuresに格納されたデータを可視化
for i, (p0,p1) in enumerate(pairs):
    ax = axes.flat[i]
    for t, marker, c in zip(range(3), ">ox", "rgb"):
        #クラスごとに色の異なるマーカでプロットする
        ax.scatter(features[target == t,p0],features[target == t,p1],marker=marker,c=c,s=40)
    #軸情報の追加
    ax.set_xlabel(features_names[p0])
    ax.set_ylabel(features_names[p1])
    #ax.set_xticks([])
    ax.set_yticks([])
```

### 1.1.2 分類モデルの作成

今回の目的が花を3つのグループに分類することであれば，いくつか方法を考える事ができる．<br>
可視化した結果を見ると，"花弁の長さ(Petal length)"を用いる事で，Setosaとその他の品種を分ける事ができそうである．<br>
以下のコードで，その境界線を見つける．

```python
#使用する情報を変数に格納
target_names = data['target_names']
labels = target_names[target]
plength = features[:, 2]

#label名が'setosa'と一致するものはTrue，一致しないものはFalseと定義して変数に格納
is_setosa = (labels == 'setosa')

#setosaの花弁の長さの最大値となる値をmax_setosaに格納
max_setosa = plength[is_setosa].max()
#他2種の花弁の長さの最小値となる値をmin_non_setosaに格納
min_non_setosa = plength[~is_setosa].min()

print('Maximum of setosa: {0}.'.format(max_setosa))
print('Minimum of others: {0}.'.format(min_non_setosa))
```

<br>
上記に示したコードの実行結果を見ると，setosaの"花弁の長さ(Petal length)"の最大値は1.9であり，他の2品種の"花弁の長さ(Petal length)"の最小値は3.0であるという事がわかる．そのため，次のような単純なモデルを作成する事ができる．<br>
それは「もし花弁の長さの値が2より小さければ，それはSetosaという品種のアイリスであり，それ以外であれば，それはVirginicaかVersicolorという品種のアイリスである」というモデルであり，以下にこのモデルを関数として定義する．

```python
def apply_model(example):
    if example[2] < 2 : print 'Iris Setosa'
    else: print 'Iris Virginica or Iris Versicolor'
```

定義した関数は単純だが，Setosaとその他のアイリスを誤り無く分類する事ができる．<br>
ただし，行った事は1つの次元(特徴量)において閾値を定めただけである．そのため，可視化結果のグラフを見ながら視覚的に分類に適している特徴量を見つけ，簡単な計算を行ったが，機械学習では以上のことを代わって行ってくれる．

Setosaとその他のアイリスを分類することが簡単であったが，VirginicaとVersicolorについては，分類のための最適な閾値を見つけるのが難しい事が見て分かる．しかし，ここではできるだけ最前の方法で分類を行えないか考える．そこで，次のような簡単な計算をまず行う．

```python
#Setosa以外の特徴量とラベルを選択し変数に格納
setosa = (labels == 'setosa')
features = features[~is_setosa]
labels = labels[~is_setosa]
virginica = (labels == 'virginica')
```

上記のコードでは，Numpyに関連する演算を多くしようしている．is_setosaはブーリアン型(真理値の真か偽の2値をとる)の配列であり，これを用いてSetosa以外の2品種のアイリスだけからなる配列(featuresとlabels)を生成できる．最後に，ラベルについての比較を行い，virginicaというブーリアン形配列を新たに作成する．<br><br>
次に，全ての特徴量とその閾値の組み合わせについての正解率(accuracy)を計算し，正解率の最も高い組み合わせを見つける．ここでの正解率とは，モデルが正しく分類できた割合を示す．また，閾値の候補は学習データの各特徴量の値を用いることにする．

```python
best_acc = -1.0

for fi in range(features.shape[1]):
    #各特徴量ごとに閾値の候補を生成する
    thresh = features[:, fi].copy()
    thresh.sort()
    #全ての閾値でテストする
    for t in thresh:
        pred = (features[:, fi] > t)
        acc = (pred == virginica).mean()
        if acc > best_acc:
            best_acc = acc
            best_fi = fi
            best_t = t
print('Best cut is {0} on feature {1}, which achieves accuracy of {2:.1%}.'.format(
    best_t, best_fi, best_acc))
```

<br>
上記のコード中のpredという変数は，各サンプルデータがvirginicaであるかどうかを示す予測結果であり，ブーリアン型の配列となる．そして，このpred配列のTrueである要素のラベル名を正解のラベル名であるvirginicaと比較を行う<br>
また，ここではブーリアン型配列の平均(mean)をとることで，ブーリアン型配列のTrueの割合を求める事ができる．このTrueの割合が，正解率の値でもあるため簡単な記述で正解率の計算が行える．forループの最後では，全ての閾値と特徴量の組み合わせで評価が行われ，より適した値がbest_fiとbest_tに格納される．以下にこのモデルを使用するための関数を定義する．

```python
def apply_model(example):
    if example[best_fi] > best_t: print 'virginica'
    else:print 'versicolor'
```

全てのデータを対象に上記のコードを実行すると，最適なモデルは"花弁の幅(Petal width)"について分類するのが最適である分かる．

下記のコードによって表示される図を見ると，2つの領域に分けられる．一つは白であり，もう一つはグレイの領域となる．
白の領域にあるものはすべてVirginicaであり，それ以外はVersicolorの品種に分類されるということを表している．<br>
閾値を用いるモデルでは，決定境界(decision boundray)は常に一つの軸に対して平行な直線になる．今回の図では縦軸に対して平行に決定境界が示されており，各点は白かグレイ領域のどちらかに属する．また，同じ正解率で分類できる閾値はもう一つ存在し，そちらはより間隔の短い点線で示している．どちらの閾値を用いるかは任意に選択できる．

```python
COLOUR_FIGURE = False

from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
data = load_iris()
features = data['data']
feature_names = data['feature_names']
species = data['target_names'][data['target']]

setosa = (species == 'setosa')
features = features[~setosa]
species = species[~setosa]
virginica = species == 'virginica'

#閾値の値をtに代入する
t = 1.75
#使用する特徴量の次元数を指定
p0, p1 = 3, 2

if COLOUR_FIGURE:
    area1c = (1., .8, .8)
    area2c = (.8, .8, 1.)
else:
    area1c = (1., 1, 1)
    area2c = (.7, .7, .7)

x0, x1 = [features[:, p0].min() * .9, features[:, p0].max() * 1.1]
y0, y1 = [features[:, p1].min() * .9, features[:, p1].max() * 1.1]

plt.figure(figsize=(8,6))
plt.fill_between([t, x1], [y0, y0], [y1, y1], color=area2c)
plt.fill_between([x0, t], [y0, y0], [y1, y1], color=area1c)
plt.plot([t, t], [y0, y1], 'k--', lw=2)
plt.plot([t - .1, t - .1], [y0, y1], 'k:', lw=2)
plt.scatter(features[virginica, p0],
            features[virginica, p1], c='b', marker='o')
plt.scatter(features[~virginica, p0],
            features[~virginica, p1], c='r', marker='x')
plt.ylim(y0, y1)
plt.xlim(x0, x1)
plt.xlabel(feature_names[p0])
plt.ylabel(feature_names[p1])
```
